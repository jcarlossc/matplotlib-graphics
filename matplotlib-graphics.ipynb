<h2>Importando a biblioteca</h2>

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

<h2>ðŸ“Œ GrÃ¡fico de Pizza</h2>
<p style='font-size:16px;'>Excelente para mostrar proporÃ§Ãµes relativas entre partes de um todo.</p>

```python
# Listas fictÃ­cias.
categorias = ['A', 'B', 'C', 'D']
valores = [40, 25, 15, 20]

plt.pie(valores, labels=categorias, autopct='%1.1f%%', colors=['#FF9999','#66B2FF','#99FF99','#FFCC99'])
plt.title("DistribuiÃ§Ã£o Percentual por Categoria")
plt.show()
```

<h2>ðŸ“Œ GrÃ¡fico de Barras</h2>
<p style='font-size:16px;'>Um dos mais usados para comparar categorias ou quantidades absolutas.</p>

```python
# Listas fictÃ­cias.
categorias = ['A', 'B', 'C', 'D']
valores = [23, 45, 12, 30]

plt.bar(categorias, valores, color='coral')
plt.title("Vendas por Categoria")
plt.ylabel("Quantidade")
plt.show()
```

<h2>ðŸ“Œ GrÃ¡fico de DispersÃ£o</h2>

<p style='font-size:16px;'>Usado para mostrar a relaÃ§Ã£o entre duas variÃ¡veis numÃ©ricas.</p>

```python
# VariÃ¡veis Numpy, fictÃ­cias.
x = np.random.rand(50)
y = x * 2 + np.random.normal(0, 0.2, 50)

plt.scatter(x, y, color='green')
plt.title("RelaÃ§Ã£o entre Duas VariÃ¡veis")
plt.xlabel("VariÃ¡vel X")
plt.ylabel("VariÃ¡vel Y")
plt.show()
```

<h2>ðŸ“Œ Boxplot (Diagrama de Caixa)</h2>
<p style='font-size:16px;'>O boxplot mostra a dispersÃ£o dos dados, destacando mediana, quartis e possÃ­veis outliers.</p>

```python
# Dados Numpy, fictÃ­cios
dados = [np.random.normal(0, std, 100) for std in range(1, 4)]

plt.boxplot(dados, vert=True, patch_artist=True)
plt.title("DistribuiÃ§Ã£o de Dados - Boxplot")
plt.xlabel("Grupos")
plt.ylabel("Valores")
plt.show()
```

<h2>ðŸ“Œ GrÃ¡fico de Densidade</h2>
<p style='font-size:16px;'>Mostra a distribuiÃ§Ã£o contÃ­nua de dados de forma suavizada</p>

```python
# Dados Numpy, fictÃ­cios
dados = np.random.normal(0, 1, 1000)

sns.kdeplot(dados, shade=True, color='purple')
plt.title("DistribuiÃ§Ã£o de Densidade")
plt.show()
```

<h2>ðŸ“Œ GrÃ¡fico de Ãrea</h2>
<p style='font-size:16px;'>O grÃ¡fico de Ã¡rea mostra a evoluÃ§Ã£o de uma variÃ¡vel acumulada, preenchendo a Ã¡rea sob a curva.</p>

```python
# VariÃ¡veis Numpy, fictÃ­cias.
x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.fill_between(x, y, color='skyblue', alpha=0.5)
plt.plot(x, y, color='steelblue')
plt.title("GrÃ¡fico de Ãrea")
plt.show()
```

<h2>ðŸ“Œ Heatmap (Mapa de Calor)</h2>
<p style='font-size:16px;'>O heatmap representa valores por meio de cores, permitindo perceber rapidamente padrÃµes e intensidades.</p>

```python
# Dados Numpy, fictÃ­cios
dados = np.random.rand(10, 10)

sns.heatmap(dados, cmap="YlGnBu")
plt.title("Mapa de Calor de CorrelaÃ§Ã£o")
plt.show()
```

<h2>ðŸ“Œ GrÃ¡fico de Facetamento (Subplots)</h2>
<p style='font-size:16px;'>No Matplotlib, o facetamento Ã© feito usando subplots, que permitem comparar dados em mÃºltiplos painÃ©is de forma organizada.</p>

```python
fig, axes = plt.subplots(1, 2, figsize=(10,4))

eixo_x = np.arange(1, 6)
eixo_y1 = [10, 20, 15, 25, 30]
eixo_y2 = [5, 15, 10, 20, 25]

axes[0].bar(eixo_x, eixo_y1, color='orange')
axes[0].set_title("Categoria A")

axes[1].bar(eixo_x, eixo_y2, color='teal')
axes[1].set_title("Categoria B")

plt.suptitle("ComparaÃ§Ã£o de Categorias com Facetamento")
plt.show()
```

<h2>ðŸ“Œ GrÃ¡fico de Linha de TendÃªncia</h2>
<p style='font-size:16px;'>O grÃ¡fico de linha Ã© o tipo mais usado para mostrar a evoluÃ§Ã£o de uma variÃ¡vel ao longo do tempo.
Ã‰ excelente para identificar tendÃªncias, padrÃµes e sazonalidades.</p>

```python
# Listas fictÃ­cias.
anos = [2018, 2019, 2020, 2021, 2022]
vendas = [150, 200, 180, 250, 300]

plt.plot(anos, vendas, marker='o', color='steelblue')
plt.title("TendÃªncia de Vendas ao Longo dos Anos")
plt.xlabel("Ano")
plt.ylabel("Vendas (em milhares)")
plt.show()
```
